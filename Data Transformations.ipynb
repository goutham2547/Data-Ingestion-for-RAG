# Data TRansformation
- convert the column which has wrong datatype to correct datatype
- convert text categorical into numeric
- convert the column, which has high magnitude to low magnitude

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Age":[23,25,30,"32",40,50,31]})
df
```

```python
df.dtypes
```

```python
df["Age"].unique()
```

```python
df["Age"] = df["Age"].astype(int)
```

# Encoding

- convert text categorical into numeric

```python
df = pd.DataFrame({"Gender":["Male","Male","female", "Male", "female","XY","ij","ij"]})
df
```

```python
df["Gender"].unique()
```

```python
df["Gender"] = df["Gender"].str.capitalize()

df["Gender"] 
```

- Label Encoding == alphabetical encoding

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df["Gender"]  = le.fit_transform(df[["Gender"]])

df
```
