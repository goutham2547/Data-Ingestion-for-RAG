```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"studentid":list(range(1001,1010)),
                    "age":[24,26,np.nan,"srk",15,18,"20",20.6,"19.4"]})
df
```

```python
df["age"].unique().tolist()
```

```python
df.columns.tolist()
```

```python
#observe the wrong data

l=[]
for i in df["age"].unique().tolist():
    if type(i) not in [int, float]:
        l.append(i)
print(l)
```

```python
#filter/select the wrong data

df[df["age"].isin(l)]
```

**treat the wrong data**

```python
df.drop(index=df[df["age"].isin(l)].index.tolist())
```

**replace**

1. replace with original value by contacting backend team (best option)

```python
#df.iloc[index]  = new values

df.iloc[[3,6,8],1] = [25,26,27]
```

2. replace with statistical value(mean for numeric, mode for categorical)

```python
df.iloc[[3,6,8],1] = df["age"].mean()
```

# Missing values

```python
df = pd.DataFrame({"studentid":[101,102,np.nan,104,105],
                    "age":[24,27,np.nan,28,np.nan]})
df
```

```python
#observe the missing values

df.isnull()
```

```python
#to observe no.of missing values in each column
df.isnull().sum()
```

```python
#to observe % of missing values in each column
df.isnull().sum()/df.shape[0]*100
```

```python
#to observe no.of missing values in each row
df.isnull().sum(axis=1)
```

```python
#to observe % of missing values in each row
df.isnull().sum(axis=1)/df.shape[1]*100
```

```python
# filter/extract the missing values
df[df["age"].isnull()]
```

**treat missing values**

1. Drop/Remove the missing values

```python
df.drop(index= df[df["age"].isnull()].index)
```

```python
df.dropna()
```

if original data is available : replace with original data

elif % of missing values is <5%  ---> drop the records

elif % of missing values is >30%  ---> drop the column

else, replace statistically

```python
df["age"] = df["age"].fillna(20)
```

# Outliers

**how to observe the outliers**

```python
df = pd.DataFrame({"X":[1,26,5,10,15,20,18,14,19,24,30,36,22,27,100,150]})
df
```

```python
Q1 = df["X"].quantile(0.25)

Q3 = df["X"].quantile(0.75)

IQR = Q3-Q1

ll = Q1-(1.5*IQR)
print("Lower Limit:",ll)

ul = Q3+(1.5*IQR)
print("Upper Limit:",ul)
```

```python
df["X"].quantile([0,0.25,0.5,0.75,1])
```

**extract outliers**

```python
df[(df["X"]<ll) | (df["X"]>ul)]
```

```python
df[(df["X"]<ll) | (df["X"]>ul)].shape[0]
```

- if outlier is wrong data, treat wrong data

        - replace with original data  (1st priority)
        - replace statically

- if outlier is correct data,

        - remove
        - retrain

```python
df.iloc[[14,15],0] = [10,15]
```
